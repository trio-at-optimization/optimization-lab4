```python
import sys
sys.path.append('../')

import numpy as np
import helper
from scipy.optimize import LinearConstraint, NonlinearConstraint
```

```python
def fun(x):
    return x[0] ** 2 + x[1] ** 2

def rosen(x):
    return (1 - x[0]) ** 2 + 100 * (x[1] - x[0] ** 2) ** 2

# Задаем линейные ограничения
x0 = (-0.0, 0.8)
x_lin = np.linspace(-1.6, 1.3, 400, dtype=float)
y_lin = np.linspace(-0.1, 1.1, 400, dtype=float)

# # Задаем границы переменных
# bounds = ((-1, 1), (-1, 1))

# Без ограничений
helper.minimize_and_output(
    rosen, x0, x_lin, y_lin,
    'trust-constr', 'trust-constr', 'trust-constr with non-linear constaints', 
    constraints=None, manual_history=True, nth=3, min_coord=(1, 1)
)


# С нелинейными ограничениями
y_lin = np.linspace(0.5, 1.1, 400, dtype=float)
nonlinear_constraints = [NonlinearConstraint(lambda x: (x[0] ** 2) / 3 + (x[1] ** 2) / 0.5 - 0.6, 0.5, np.inf)]
helper.minimize_and_output(
    rosen, x0, x_lin, y_lin,
    'trust-constr', 'trust-constr', 'trust-constr with non-linear constaints', 
    constraints=nonlinear_constraints, manual_history=True, nth=3, min_coord=(1, 1)
)


# С линейными ограничениями
y_lin = np.linspace(0.6, 1.3, 400, dtype=float)
linear_constraints = [LinearConstraint(np.array([0.1, 1]), lb=0.78)]
helper.minimize_and_output(
    rosen, x0, x_lin, y_lin,
    'trust-constr', 'trust-constr', 'trust-constr with non-linear constaints',
    constraints=linear_constraints, manual_history=True, nth=3, min_coord=(1, 1)
)
```

```python
def f_tanga(x, _=None):
    return sum([xi ** 4 - 16 * xi ** 2 + 5 * xi for xi in x])


x0 = (-4.0, -4.0)
x_lin = np.linspace(-5, 5, 1000, dtype=float)
y_lin = np.linspace(-5, 5, 1000, dtype=float)


# Без ограничений
helper.minimize_and_output(
    f_tanga, x0, x_lin, y_lin,
    'trust-constr', 'trust-constr', 'trust-constr with non-linear constaints', 
    constraints=None, bounds=None, manual_history=True, min_coord=(-2.9, -2.9)
)

# С нелинейными ограничениями
nonlinear_constraints = [NonlinearConstraint(lambda x: (x[0] + 3) ** 2 + (x[1] + 3) ** 2, 0.25, np.inf)]
helper.minimize_and_output(
    f_tanga, x0, x_lin, y_lin,
    'trust-constr', 'trust-constr', 'trust-constr with non-linear constaints', 
    constraints=linear_constraints, manual_history=True, nth=3, min_coord=(-2.9, -2.9)
)
```
